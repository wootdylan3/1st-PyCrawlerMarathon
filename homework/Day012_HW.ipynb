### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, "[" + d.find(class_="tag").text + "]", d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# use dictionary
newslist = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    tag = d.find(class_="tag").text
    newslist.setdefault(tag, [])
    newslist[tag].append({
        'date': date,
        'title': title        
    })
    
print(newslist)
```

### ③ 哪一個類別的文章最多

```python
newsmax = 0
itemmax = ''
for item in newslist:
    print(item, len(newslist[item]), "篇")
    if (len(newslist[item]) > newsmax):
        newsmax = len(newslist[item])
        itemmax = item
print("文章最多的類別:", itemmax, "共有", len(newslist[itemmax]), "篇")
    
```
